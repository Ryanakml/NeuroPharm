Bayangin kita punya rumah yang pintunya rusak, dan pintu itu bikin maling gampang masuk. Nah, pintu itu di sini kita ibaratkan protein target, yaitu bagian dalam tubuh yang kalau aktif atau berubah bentuk, bisa bikin penyakit muncul misalnya kanker atau virus.

Sekarang kita pengen mengunci pintu itu supaya maling gak bisa masuk lagi. Tapi, kita gak bisa pakai sembarang kunci. kita perlu kunci yang pas banget untuk pintu itu. Nah, kunci ini kita ibaratkan molekul obat. Molekul ini harus pas dan kuat nempel di protein target supaya bisa “menonaktifkan” atau “menghambat” protein tersebut agar penyakit nggak jalan.

Tapi, masalahnya, ada jutaan kunci (molekul) yang tersedia. Kalau kita coba satu-satu kunci itu secara manual, kita butuh waktu dan tenaga yang sangat besar. Ini sama seperti di dunia nyata: para peneliti harus uji ribuan molekul di laboratorium untuk cari yang pas. Proses ini mahal dan lama, bisa sampai bertahun-tahun.

*Nah, ini kenapa kita pakai AI. AI di sini berfungsi seperti mesin pintar yang sudah belajar dari kunci-kunci dan pintu yang pernah diuji sebelumnya. Jadi, AI bisa “menebak” kunci mana yang kemungkinan besar cocok dengan pintu tertentu, tanpa harus uji satu-satu secara fisik. Ini menghemat waktu dan biaya secara drastis.*

#### AI and Medicine

Untuk bisa bikin AI pintar seperti itu, pertama-tama kita butuh data. Data ini datang dari dua sumber utama: Protein Data Bank (PDB) yang berisi informasi tentang struktur 3D protein target (pintu kita), dan ChEMBL yang berisi data tentang molekul-molekul yang pernah diuji dan bagaimana aktivitasnya terhadap protein tertentu (kunci-kunci yang sudah pernah dicoba).

Tapi data ini mentah dan sering kacau. Jadi, kita harus bersihkan dan rapikan dulu supaya AI bisa ngerti. Misalnya, struktur protein yang asli dalam bentuk 3D harus diubah supaya komputer bisa pahami, dan molekul-molekul juga harus diubah jadi format yang bisa diproses, seperti “sidik jari” angka yang mewakili bentuk dan sifat molekul tersebut.

Selanjutnya, kita buat AI yang bisa belajar hubungan kompleks antara kunci dan pintu. Model yang kita pakai namanya Graph Neural Network (GNN), yang jago banget memahami hubungan antar atom dalam molekul (kunci) dan pola interaksinya dengan protein (pintu). Jadi, AI ini belajar bukan cuma dari data biasa, tapi dari struktur kimia yang rumit.

Tapi prediksi AI itu sifatnya matematis dan statistik, belum tentu 100% akurat secara fisik. Makanya kita lakukan simulasi docking, yang seperti uji coba virtual di komputer untuk melihat bagaimana si kunci (molekul) bisa masuk dan menempel di pintu (protein) secara realistis. Ini pakai software bernama AutoDock Vina yang ngitung posisi terbaik dan kekuatan ikatan molekul-protein secara fisik.

Hasil dari AI dan simulasi ini kita kumpulkan di sebuah platform, yang gampang diakses oleh peneliti lewat dashboard web. Di sana, mereka bisa lihat molekul mana yang paling menjanjikan, lengkap dengan visualisasi 3D supaya mereka benar-benar paham bagaimana molekul menempel ke protein. Jadi, gak perlu repot coding atau utak-atik data mentah.

Teknologi yang kita pilih itu seperti alat-alat andalan. RDKit untuk olah molekul, Biopython untuk protein, PyTorch Geometric untuk bikin AI yang bisa paham grafik molekul, ESM-2 embedding buat ringkas protein jadi data angka bermakna, AutoDock Vina buat simulasi fisika, FastAPI buat bikin API backend supaya semua fungsi bisa dipakai dengan mudah, dan Streamlit untuk bikin dashboard interaktif.

Intinya, kita bikin alat pintar dan lengkap yang mulai dari data mentah sampai hasil visual dan prediksi bisa diakses gampang dan cepat. Ini bisa memangkas proses riset obat yang biasanya ribet dan mahal, jadi lebih efektif dan efisien.
