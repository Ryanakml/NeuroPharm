# NeuroPharm - AI Drug Discovery

## Section 1 : Problem Statement

### Why?

Bayangin proses penemuan obat itu kayak nyari pasangan kunci sama gembok. Gemboknya adalah protein penyebab penyakit, dan kuncinya adalah molekul kecil yang bisa nyangkut ke protein itu dengan pas, sehingga mengganggu cara kerjanya dan mencegah penyakitnya jalan. Masalahnya, jumlah kunci yang mungkin itu gila banyaknya—bisa jutaan bahkan miliaran bentuk. Kalau mau nyoba satu-satu secara fisik di lab, prosesnya bisa makan waktu 10-15 tahun dan biaya sampai miliaran dolar. Dan sering kali, kuncinya ternyata nggak pas, walaupun di awal kelihatan cocok.

Di dunia farmasi, ini disebut High-Throughput Screening (HTS)—menguji ribuan sampai jutaan molekul secara fisik untuk lihat yang mana yang “nyantol”. Tapi ini kayak nyari jarum di tumpukan jerami sambil matanya ditutup. Banyak yang ketemu, tapi ternyata cuma “mirip jarum” dan nggak bisa dipakai. Nah, di sinilah AI masuk sebagai “detektor jarum pintar”. Alih-alih kita bongkar jerami sampai lelah, AI bisa memindai secara virtual semua kandidat, mempelajari pola dari data sebelumnya, lalu memprediksi mana yang paling mungkin jadi kunci yang beneran pas.

### AI Solution

Kalau cara lama itu nyaring dari jutaan sampai dapat segelintir, AI bisa membalik prosesnya: kita langsung fokus ke ratusan atau ribuan kandidat terbaik, lalu baru diuji di lab. Ini bisa memangkas waktu riset awal hingga separuhnya dan meningkatkan peluang sukses di tahap uji klinis yang super mahal. Dan sekarang, AI bukan cuma sekadar nyaring molekul yang udah ada, tapi dengan teknologi seperti Graph Neural Networks (GNNs), dia bisa “paham” logika kimia dan fisika yang bikin kunci-gembok itu cocok. Artinya, ke depannya, AI bahkan bisa mendesain kunci baru yang sebelumnya nggak ada—ini yang disebut de novo drug design.

Nah, platform yang kita bangun ini sekarang fokusnya masih di tahap penyaringan (virtual screening), tapi pondasinya kita bikin supaya nanti bisa dipakai juga buat generasi molekul baru. Anggap aja sekarang kita masih jadi “pencari kunci tercepat di dunia”, tapi pelan-pelan kita bakal naik level jadi “arsitek kunci masa depan”.

### Pipelines

Dari sisi teknis, sistem kita ini jalannya lewat tiga tahap besar. Pertama, kita ngumpulin dan nyiapin data. Data protein target kita ambil dari Protein Data Bank (PDB), yang nyimpen struktur 3D protein hasil riset dari seluruh dunia. Data molekul calon obat dan seberapa aktif dia (bioaktivitas) kita ambil dari ChEMBL. Masalahnya, data ini sering berantakan, formatnya beda-beda, kadang ada yang hilang. Jadi sebelum AI bisa belajar, kita harus bersihin dulu datanya, rapihin, dan standarisasi. Di tahap ini juga kita bakal ngelakuin Exploratory Data Analysis (EDA), semacam cek kondisi medan perang—ngeliat sebaran data, pola, anomali, dan potensi masalah yang bisa ngeganggu model nanti.

Tahap kedua adalah inti AI-nya. Di sini, kita harus mengubah molekul dan protein jadi angka—karena komputer cuma ngerti angka, bukan bentuk atau huruf. Untuk molekul, kita pakai fingerprints dan representasi graf. Fingerprints ini kayak sidik jari digital yang menggambarkan ciri khas molekul. Representasi graf lebih canggih lagi—molekul dianggap sebagai peta hubungan antara atom (node) dan ikatan (edge). Ini penting karena bentuk dan koneksi atom itu pengaruh banget ke cara dia berikatan sama protein. Untuk protein, kita pakai model bahasa protein seperti ESM-2 yang bisa bikin “embedding” alias representasi numerik dari urutan asam amino, jadi AI bisa ngerti karakteristiknya tanpa harus selalu punya struktur 3D-nya. Semua angka ini jadi bahan bakar untuk melatih model AI yang tugasnya memprediksi binding affinity—seberapa kuat si molekul bakal nempel di protein target.

Tahap ketiga adalah validasi docking. Walaupun AI udah memprediksi, itu cuma prediksi statistik. Kita butuh “opini kedua” dari sisi fisika. Docking adalah simulasi komputer yang ngecek apakah molekul bener-bener bisa nyangkut di protein secara realistis. Kita pakai AutoDock Vina yang terkenal cepat dan akurat untuk ini. Hasilnya adalah skor berbasis energi, makin rendah energinya, makin stabil ikatannya. Jadi kita bisa bandingin prediksi AI dan hasil docking untuk yakin bahwa kandidat kita nggak cuma bagus di teori tapi juga di simulasi fisik. Semua ini nantinya kita tampilkan di dashboard web interaktif, lengkap dengan visualisasi 3D protein dan molekul, skor, dan ranking—supaya peneliti tinggal klik-klik buat ambil keputusan.

### Tech Included

Pemilihan teknologi yang tepat adalah kunci keberhasilan proyek ini, apalagi di ranah bioinformatika yang memadukan data kimia dan biologi dengan skala besar dan kompleksitas tinggi. Seluruh stack yang digunakan bersifat open-source, sudah teruji di industri, dan memiliki komunitas yang aktif, sehingga memudahkan pengembangan sekaligus memastikan dukungan jangka panjang.

Untuk analisis molekul, kita menggunakan RDKit, sebuah library serbaguna dalam cheminformatics. RDKit memungkinkan kita membaca, menulis, dan memanipulasi struktur kimia dari berbagai format seperti SMILES. Dari sini, kita bisa menghitung descriptors seperti berat molekul atau nilai LogP, dan menghasilkan fingerprints seperti Morgan Fingerprints yang akan merepresentasikan molekul dalam bentuk numerik sebagai input untuk model machine learning. Keunggulan RDKit adalah sifatnya yang fleksibel, gratis, dan kompatibel dengan ekosistem data science Python seperti Pandas dan Scikit-learn.

Sementara itu, untuk data protein, kita memanfaatkan Biopython, toolkit fundamental untuk bioinformatika. Modul Bio.PDB di dalamnya memudahkan kita mengunduh struktur protein dari database PDB, mem-parsing data tersebut, mengekstrak rantai asam amino, serta memeriksa hierarki strukturnya hingga tingkat atom. Fungsi ini membuat kita bisa mengelola dan menganalisis data protein secara sistematis.

Meski fingerprints molekul bermanfaat, bentuk representasi ini mereduksi molekul menjadi vektor datar sehingga menghilangkan informasi topologi 3D yang penting. Untuk mempertahankan informasi tersebut, kita memanfaatkan PyTorch Geometric (PyG) yang dirancang untuk memproses data berbentuk graf. Dengan ini, molekul dapat direpresentasikan sebagaimana bentuk aslinya, yaitu atom sebagai node dan ikatan sebagai edge, sehingga memungkinkan kita membangun Graph Neural Networks (GNN) yang memahami struktur molekul secara lebih utuh.

Pendekatan serupa kita terapkan pada protein menggunakan model ESM-2 dari Meta AI, sebuah Protein Language Model (PLM) yang bekerja mirip dengan model bahasa seperti GPT. Model ini dilatih pada jutaan sekuens protein dan mampu menghasilkan embeddings yang mengandung informasi evolusioner, struktural, dan fungsional hanya dari sekuens asam amino. Dengan bantuan library transformers dari Hugging Face, kita dapat langsung mengakses model ini dan memanfaatkan representasi yang dihasilkannya tanpa harus memiliki data struktur 3D eksperimental.

Untuk memastikan hasil prediksi AI kita selaras dengan realitas fisika, kita menggunakan AutoDock Vina, sebuah program molecular docking yang populer di akademisi maupun industri. Vina bekerja dengan mencari pose ikatan ligan paling stabil di dalam binding pocket protein dan menghitung skor afinitas yang berkorelasi dengan energi bebas ikatan. Kecepatan dan akurasi Vina menjadikannya pilihan ideal untuk tahap validasi.

Dalam tahap deployment, kita menggabungkan FastAPI sebagai backend dan Streamlit sebagai frontend. FastAPI memberikan cara cepat dan modern untuk membangun API yang dapat diakses aplikasi lain, lengkap dengan validasi otomatis dan dokumentasi interaktif bawaan. Sementara itu, Streamlit memungkinkan pembuatan dashboard web interaktif hanya dengan Python, sehingga kita bisa menampilkan hasil prediksi, visualisasi molekul, atau grafik performa model secara real-time tanpa perlu menulis kode JavaScript atau HTML.

Dengan kombinasi teknologi ini, kita memiliki alur kerja yang utuh: mulai dari pengolahan molekul dan protein, representasi canggih berbasis AI, validasi berbasis fisika, hingga deployment dalam bentuk aplikasi interaktif yang siap digunakan peneliti atau tim pengembang obat.

Kalau ini kita bikin dengan rapi dari awal, nanti nambah fitur kayak generasi molekul baru atau integrasi lab otomatis bakal jauh lebih gampang. Jadi, ini bukan cuma proyek AI, tapi pondasi buat ekosistem penemuan obat yang bisa terus berkembang.

```python
def download_protein_structure(pdb_id, download_dir='data/pdb'):
    """
    Mengunduh file PDB untuk PDB ID tertentu.
    """
    os.makedirs(download_dir, exist_ok=True)
    pdbl = PDBList()
    pdb_file_path = pdbl.retrieve_pdb_file(pdb_id, pdir=download_dir, file_format='pdb', overwrite=True)

    if not os.path.exists(pdb_file_path):
        print(f"Gagal mengunduh file PDB untuk {pdb_id}.")
        return None

    print(f"Protein Structure {pdb_id} successfully downloaded to: {pdb_file_path}")
    return pdb_file_path

# Tentukan PDB ID target kita
TARGET_PDB_ID = '2GS2'
pdb_file = download_protein_structure(TARGET_PDB_ID)
```
